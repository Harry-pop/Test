```python
import matplotlib.pyplot as plt
x = [5,6,7,8,9]
y =['Mon','Tue','Wed','Thr','Fri']
plt.style.use('seaborn')
plt.title('Graph',color='r',fontsize=14)
plt.xlabel('x',color='b',fontsize=10)
plt.ylabel('y',color='g',fontsize=10)
plt.xticks(rotation='vertical',color='k')
plt.yticks(color='r')
plt.plot(x,y)
plt.show()
```
